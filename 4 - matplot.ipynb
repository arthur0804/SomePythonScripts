```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# Funtional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# subplot (rows, cols, plot number)
plt.subplot(1,2,1) # or (121)
plt.plot(x,y,'r')

plt.subplot(1,2,2) # or (122)
plt.plot(y,x,'b')

plt.show()
```

```python
# Object Oriented method
fig = plt.figure()

# four arguments: left & bottom (position relation to the canvas); width & height
# (all from 0 to 1)
axes = fig.add_axes([0.1 , 0.1 , 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# subplot
fig = plt.figure()

axes1 = fig.add_axes([0.1 , 0.1 , 0.8, 0.8])
axes2 = fig.add_axes([0.2 , 0.5 , 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger')
axes2.plot(y,x)
axes2.set_title('Smaller')
```

```python
# create fig and axes at the same time
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
    
plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(5,3))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

### Legends

```python
fig = plt.figure(figsize=(5,3))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X cubed')

# add legend
# use "loc=" to choose location
ax.legend(loc=0)
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# use "color=" to set colors
# use "linewidth/lw=" to set linewidth
# use "alpha=" to set transparency
# use "linestyle/ls=" to set style
# use "marker=" to set markers of data points
ax.plot(x,y,color='red', linewidth=2, alpha=0.5, linestyle='--', marker='o', markersize=5)

# set the scale
ax.set_xlim([0,10])
ax.set_ylim([0,10])
```

### Data Points with annotations

```python
plt.axis([0,5,0,20]) 
plt.title('My first plot',fontsize=20,fontname='Times New Roman') 
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First') 
plt.text(2,4.5,'Second') 
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

### Adding a grid

```python
plt.axis([0,5,0,20]) 
plt.title('My first plot',fontsize=20,fontname='Times New Roman') 
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First') 
plt.text(2,4.5,'Second') 
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid(True)
plt.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

### Other plot types

### 1. Line charts

```python
import numpy as np
import pandas as pd
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(3*x)/x
plt.plot(x,y)
plt.show()
```

```python
data = {'series1':[1,3,4,3,5], 
        'series2':[2,4,5,2,4], 
        'series3':[3,2,3,1,3]} 
df = pd.DataFrame(data)
x = np.arange(5) 
plt.axis([0,4,0,7])
plt.plot(x,df,marker='o') 
plt.legend(data, loc=0)
plt.show()
```

### 2. Histograms

```python
pop = np.random.randint(0,100,100)
pop
```

```python
# use "bins=" to specify how many bins to be divided into 
plt.hist(pop,bins=20)
plt.show()
```

### 3. Bar Charts

```python
index = [0,1,2,3,4] 
values = [5,7,3,4,6]
# use "width=" to specify width between bars
plt.bar(index,values, width=0.5)
plt.show()
```

### 4. Pie Chart

```python
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors = ['yellow','green','red','blue']
plt.pie(values,labels=labels,colors=colors)
plt.axis('equal')
plt.show()
```

### 5. Scatter plot

```python
x = np.random.randint(0,10,10)
y = np.random.randint(10,20,10)
plt.scatter(x, y)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
